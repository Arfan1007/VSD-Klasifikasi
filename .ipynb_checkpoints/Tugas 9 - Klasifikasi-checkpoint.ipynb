# Klasifikasi


**Metode KNN**

```python
X = df.drop(columns='Date')
X.head()
```

```python
y = df['Volume']
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=10)

print('training_dataset')
print(X_train.shape)
print(y_train.shape)
print()
print('testing_dataset')
print(X_test.shape)
print(y_test.shape)
```

```python
k_range = list(range(1,26))
scores = []
for k in k_range:
    model_knn = KNeighborsClassifier(n_neighbors=k)
    model_knn.fit(X_train, y_train)
    y_pred = model_knn.predict(X_test)
    scores.append(accuracy_score(y_test, y_pred))
```

```python
plt.plot(k_range, scores)
plt.xlabel('Nilai KNN')
plt.ylabel('Akurasi Skor')
plt.title('Akurasi Tesla menggunakan KNN')
plt.tight_layout()
plt.show()
```

```python
model_knn = KNeighborsClassifier(n_neighbors=3)
model_knn.fit(X_train, y_train)
y_pred = model_knn.predict(X_test)
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

**Logistic Regressions**

```python
model_lr = LogisticRegression(solver='lbfgs', multi_class='auto')
model_lr.fit(X_train, y_train)
y_pred = model_lr.predict(X_test)
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

**SVC**

```python
model_svc = SVC(gamma='scale')
model_svc.fit(X_train, y_train)
y_pred = model_svc.predict(X_test)
```

```python
print(accuracy_score(y_test, y_pred))
```

**Random Forest**

```python
model_rf = RandomForestClassifier(n_estimators=100)
model_rf.fit(X_train, y_train)
y_pred = model_rf.predict(X_test)
```

```python
print(accuracy_score(y_test, y_pred))
```

**Decisons Tree**

```python
model_dt = DecisionTreeClassifier()
model_dt.fit(X_train,y_train)
y_pred = model_dt.predict(X_test)
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
models = [model_knn, model_lr, model_svc, model_dt, model_rf]
accuracy_scores = []
for model in models:
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)

print(accuracy_scores)
```

```python
plt.bar(['KNN', 'LogisticReg', 'SVC', 'DT', 'RF'], accuracy_scores)
plt.ylim(0.90,1.01)
plt.title('Komparasi nilai dari hasil beberpa model klasifikasi', fontsize=13, color='b')
plt.xlabel('Metode Klasifikasi', fontsize=12, color='g')
plt.ylabel('Akurasi Skor', fontsize=12, color='g')
plt.tight_layout()
plt.show()
```
